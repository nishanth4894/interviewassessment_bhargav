1. Write a program that takes a string as input, and counts the frequency of each word in the string, there might
be repeated characters in the string. Your task is to find the highest frequency and returns the length of the
highest-frequency word.

```python
def highest_frequency_word(input_string):
    words = input_string.replace(",","")
    words = input_string.split()
    word_count = {}
    for word in words:
        if word in word_count:
            word_count[word] = word_count[word] + 1
        else:
            word_count[word] = 1
        most_frequent_word = max(word_count.keys())
        most_frequent_word_length = len(max(word_count.keys()))
        
    return most_frequent_word_length, most_frequent_word
```

```python
# 1st test case
string = "write write write all the number from from from 1 to 100"
print(f"length of the highest frequency word:",highest_frequency_word(string)[0])
print(f"highest frequency word:",highest_frequency_word(string)[1])
```

#Question 2: -
Consider a string to be valid if all characters of the string appear the same number of times. It is also valid if
he can remove just one character at the index in the string, and the remaining characters will occur the same
number of times. Given a string, determine if it is valid. If so, return YES , otherwise return NO .

Note - You have to write at least 2 additional test cases in which your program will run successfully and provide
an explanation for the same.

Example input 1 - s = “abc”. This is a valid string because frequencies are { “a”: 1, “b”: 1, “c”: 1 }

Example output 1- YES

Example input 2 - s “abcc”. This string is not valid as we can remove only 1 occurrence of “c”. That leaves
character frequencies of { “a”: 1, “b”: 1 , “c”: 2 }

Example output 2 - NO

```python
def valid_string(s):
    
    char_freq = {}
    for char in s:
        char_freq[char] = char_freq.get(char, 0) + 1

    freq_freq = {}
    for freq in char_freq.values():
        freq_freq[freq] = freq_freq.get(freq, 0) + 1

    if len(freq_freq) == 1:
        return "YES"

    if len(freq_freq) > 2:
        return "NO"

    freq_list = list(freq_freq.items())
    freq1, count1 = freq_list[0]
    freq2, count2 = freq_list[1]

    if (freq1 == 1 and count1 == 1) or (freq2 == 1 and count2 == 1):
        return "YES"
    if (freq1 - freq2 == 1 and count1 == 1) or (freq2 - freq1 == 1 and count2 == 1):
        return "YES"

    return "NO"


# Test case 1
s1 = "abcde"
print(valid_string(s1))


# Test case 2
s2 = "aabbc"
print(valid_string(s2))
```

Question 10 -
Write a program to count the number of verbs, nouns, pronouns, and adjectives in a given particular phrase or
paragraph, and return their respective count as a dictionary.

Note -

1. Write code comments wherever required for code

2. You have to write at least 2 additional test cases in which your program will run successfully and provide
an explanation for the same.

```python
import nltk
from nltk import pos_tag 
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
from nltk.tokenize import word_tokenize
```

```python
def count_tags(text):
    words = word_tokenize(text)
    
    tagged_words = pos_tag(words)
    
    # Initialize for verbs, nouns, pronouns, and adjectives
    verb_count = 0
    noun_count = 0
    pronoun_count = 0
    adjective_count = 0
    
    # Iterate over the tagged words and count the occurrences of each part of speech
    for word, tag in tagged_words:
        if tag.startswith('V'):  # Verb
            verb_count += 1
        elif tag.startswith('N'):  # Noun
            noun_count += 1
        elif tag == 'PRP':  # Pronoun
            pronoun_count += 1
        elif tag.startswith('J'):  # Adjective
            adjective_count += 1
    
    # Create a dictionary with the counts
    counts = {
        'verbs': verb_count,
        'nouns': noun_count,
        'pronouns': pronoun_count,
        'adjectives': adjective_count
    }
    
    return counts

# Example usage
text = "The quick brown fox jumps over the lazy dog. It is a sunny day."
result = count_tags(text)
print(result)
```
